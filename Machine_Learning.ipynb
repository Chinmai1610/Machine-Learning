Support Vector Machine

```python
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
# 2D features (eg. ear size ,eyes , skin color )
X = np.array([[1,2],[1.5,1.8],[2,2.5], # Cats
                [4,4], [4.4,4.2], [5,5]]) # Dogs

Y = np.array([0,0,0,1,1,1]) # 0 = Cat , 1 = Dog
```

```python
# Train the model with SVM with RBF kernel

model = SVC(kernel="rbf", probability = True).fit(X,Y)
model
```

```python
print("Prediction for [3,3]:", "Dog" if model.predict([[4.6,4.9]])[0] else "Cat")
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,cmap=plt.cm.Paired)
plt.title("Support Vector Machine")
plt.xlabel("Ear Size")
plt.ylabel("Eyes Size")
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,cmap="coolwarm")
plt.title("Support Vector Machine")
plt.xlabel("Ear Size")
plt.ylabel("Eyes Size")
plt.show()
```

```python
xx,yy = np.meshgrid(np.linspace(0,6,200),np.linspace(0,6,200))
grid = np.c_[xx.ravel(),yy.ravel()]
Z = model.predict(grid).reshape(xx.shape)

plt.contourf(xx,yy,Z,cmap="coolwarm",alpha=0.5)
plt.xlabel("Ear Size")
plt.ylabel("Eyes Size")
plt.show()
```

*   K-Means Clustering



```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42) # Data - I am taking of arrays or random numbers - it would be same for everyone
```

```python
cluster1 = np.random.randn(80,2) + np.array([40,5]) # People who spend low but frequently visits
cluster2 = np.random.randn(80,2) + np.array([70,2]) # People who spend High but rarely visits
cluster3 = np.random.randn(80,2) + np.array([20,1]) # People who spend very low , rarely visits
```

```python
X = np.vstack((cluster1,cluster2,cluster3)) # Combining all the clusters
```

```python
len(X)
```

```python
kmeans = KMeans(n_clusters=3, n_init=10) # run K means 10 times with different initial centres
labels = kmeans.fit_predict(X) # runs training and return the cluster
```

```python
kmeans
```

```python
labels
```

```python
wcss = []
k_range = range(1,9)

for k in k_range:
  kmeans = KMeans(n_clusters=k, n_init=10)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
print("Before Elbow Method")
for k,val in zip(k_range,wcss):
  print(f"K = {k} , WCSS = {val}")
```

```python
print(k_range)
wcss
```

```python
drops = []
for i in range(1,len(wcss)):
  drop = wcss[i-1] - wcss[i]
  drops.append(drop)
elbow_k = np.argmax(drops) + 1

print("After Elbow Method")
print(f"Elbow K = {elbow_k}")
```

```python
k_before = 3
model_before = KMeans(n_clusters = k_before, n_init=10)
labels_before = model_before.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels_before)
plt.title("Clusters before Elbow Method")
plt.xlabel("Feature 1")
plt.xlabel("Feature 2")
plt.show()
```

```python
cluster1 = np.random.randn(80,2) + np.array([40,5]) # People who spend low but frequently visits
cluster2 = np.random.randn(80,2) + np.array([70,2]) # People who spend High but rarely visits
cluster3 = np.random.randn(80,2) + np.array([20,1]) # People who spend very low , rarely visits
```

```python
intertias = []
for k in range(1,10):
  km=KMeans(n_clusters=k,n_init=10)
  km.fit(X)
  intertias.append(km.inertia_)

plt.plot(range(1,10),intertias,marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Intertia')
plt.show()
```

```python
k_after = 5
model_after = KMeans(n_clusters = k_after, n_init=10)
labels_after = model_after.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels_after)
plt.title(f"Clusters after Elbow Method k={k_after}")
plt.xlabel("Feature 1")
plt.xlabel("Feature 2")
plt.show()
```

Euclidean and Manhattan Distance

```python
# Creating 2 simple points

import numpy as np
import matplotlib.pyplot as plt

# Two simple 2D points

p = np.array((2, 3))
q = np.array((8, 7))

# Euclidean Distance
euclidean = np.sqrt((q[0] - p[0])**2 + (q[1] - p[1])**2)

# Manhattan Distance
manhattan = np.abs(q[0] - p[0]) + np.abs(q[1] - p[1])

print("Euclidean Distance", euclidean)
print("Manhattan Distance", manhattan)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(p[0], p[1], color='red', label = "Point P (2,3)")
plt.scatter(q[0], q[1], color='blue', label = "Point Q (8,7)")

plt.plot([p[0],q[0]], [p[1], q[1]], color = "green", label = "Euclidean Path")

plt.plot([p[0],q[0],q[0]], [p[1], q[1], q[1]], color = "orange", linestyle = '--',label = "Manhattan Path")

plt.title("Euclidean vs Manhattan Distance")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Create a grid of points
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)

# Compute both distances from point p
euclid_grid = np.sqrt((X - p[0])**2 + (Y - p[1])**2)
manhattan_grid = abs(X - p[0]) + abs(Y - p[1])

# Plot side-by-side comparison
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(euclid_grid, extent=[0,10,0,10], origin='lower')
plt.title("Euclidean Distance Field")
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(manhattan_grid, extent=[0,10,0,10], origin='lower')
plt.title("Manhattan Distance Field")
plt.colorbar()

plt.show()
```

**Customer Segmentation using Unsupervised Machine Learning**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
```

```python
# Generate the Data for 300 Customers

n = 300

ages = np.random.normal(35 , 10, n).astype(int)
income = np.random.normal(600000,150000,n).astype(int)
spending_score = np.random.uniform(1,100,n).astype(int)
```

```python
df = pd.DataFrame({'Age':ages, 'Annual_Income':income, 'Spending Score':spending_score})
```

```python
df.head()
```

```python
# Preprocess the Data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
#scaled_data
```

```python
kmeans = KMeans(n_clusters=4,random_state=42)
df["Cluster"] = kmeans.fit_predict(scaled_data)

print(kmeans.cluster_centers_)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Annual_Income'], df['Spending Score'],
            c=df['Cluster'], s=50)
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation (K-Means)")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['Age'], df['Annual_Income'], df['Spending Score'],
           c=df['Cluster'], s=50)

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
ax.set_title("3D Customer Segmentation")
plt.show()
```
